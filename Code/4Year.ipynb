```python
def modify(path):
    
    df = pd.read_csv(os.path.join('Dataset_4year/', path))            
    
    df['RETURN_PRICE'] = np.log(df['Close']) - np.log(df['Close'].shift(1))
    
    df.to_csv(os.path.join('Modified_Dataset_4Year/', path), columns= df.columns)
    
    plt.plot(df['RETURN_PRICE'])
    plt.xlabel('Day')
    plt.ylabel('Return Price')
    plt.ylim(-0.3,0.3)
    plt.title(path)
    plt.show()   
```

```python
def compute_change(file, W):
    
    for i in range(4):
    
        df_new = pd.read_csv(os.path.join('Modified_Dataset_4Year/', file), index_col= 0)


        s = df_new['Adj Close'][0]

        e = np.mean(df_new['Adj Close'][i*W : (i+1)*W])

        val = (e-s)/s * 100
        
        print(e)
        
        print(f'{file[:-4]} --> {val} --> {i+2018}')
        
        if i == 0:
            year_2018.append(val)
            
        if i == 1:
            year_2019.append(val)
    
        if i == 2:
            year_2020.append(val)
    
        if i == 3:
            year_2021.append(val)
            
        print('----------')
    
    
```

```python
def colours(year):
    
    array = []
    
    for val in df_cumulative[year]:
        if val>= 0:
            array.append('Green')

        elif val>= -20:
            array.append('White')

        else:
            array.append('Red')
            
    return array
    
```

```python
matrix = pd.DataFrame()
```

```python
def build_data_matrix(path, matrix):
    
    df = pd.read_csv(os.path.join('Modified_Dataset_4Year/', path), index_col= 0)
    matrix[path[:-4]] = df['RETURN_PRICE']
    
    return matrix    
```

```python
for file in files:
    matrix = build_data_matrix(file, matrix)
```

```python
W = (len(df['RETURN_PRICE'])-1)//4
print(W)
```

```python
matrix = matrix.iloc[1:, :]
```

```python
matrix.info()
```

```python
windows = []
```

```python
for i in range(0, len(df['RETURN_PRICE']-1), W):
    window = pd.DataFrame()
    window = matrix.iloc[i:i+W]
    windows.append(window)
```
